# Crea tu Propio Modelo en Ollama usando Open-source 'Large Language Models' (LLMs).
## RI AI + Python en español (youtube)

Vídeo: https://www.youtube.com/watch?v=yR2tcdcXHSE

## Creando modelos

```python
modelfile = '''
FROM mistral-nemo
SYSTEM Your are Mister Potato
'''

ollama.create(model="mister-potato", modelfile=modelfile)
```

```python
response = ollama.chat(model="mister-potato", messages=[
    {
        "role": "user",
        "content": "Hazme una tabla de ejercicios para entrenar una carrera de 1 kilómetro",
    },
])
print(response["message"]["content"])
```

```python
response = ollama.chat(model="peter-pan", messages=[
    {
        "role": "user",
        "content": "Why don't you want to grow up?, in spanish",
    },
])
print(response["message"]["content"])
```

## Otro modelo

```python
modelfile = '''
FROM mistral
SYSTEM Your are Music Expert
'''

ollama.create(model="music-expert", modelfile=modelfile)
```

```python
response = ollama.chat(model="music-expert", messages=[
    {
        "role": "user",
        "content": "Tell me something about the music group The Cure and the name of its members, in spanish",
    },
])
print(response["message"]["content"])
```

# OLLAMA : ¿Qué son los embeddings?, ¿Qué es un embedding en IA? Crea tus embeddings con Ollama 
## RI AI + Python en español (youtube)

Vídeo: https://www.youtube.com/watch?v=PTaQQAtxkFQ

Son representaciones numéricas de las palabras. Se hace en forma de vectores (de ahí el uso de bbdd vectoriales)

```python
embedding_1 = ollama.embeddings(model="llama2", prompt="The color of the sea is blue")
```

```python
# Se imprimen los 5 primeros elementos del diccionario que se crean
print(embedding_1["embedding"][:5])
print(len(embedding_1["embedding"]))
```

Según los embeddings (sus valores) la proximidad de ellos es el que permite al modelo devolver un texto como respuesta a nuestras consultas.
